```python
import matplotlib.pyplot as plt
import numpy as np
# Creating x axis with range and y axis with Sine
# Function for Plotting Sine Graph
x = np.arange(0, 5*np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values for x between 0 and 2π
x = np.linspace(0, 2 * np.pi, 1000)

# Sine function with phase shift of 60 degrees (pi/3 radians)
y = np.sin(x + np.pi / 3)

# Create the plot
plt.plot(x, y, label="Sine function with phase shift 60°")

# Add labels and title
plt.title('Sine Function with Phase Difference of 60°')
plt.xlabel('x (radians)')
plt.ylabel('y')

# Show grid
plt.grid(True)

# Show the legend
plt.legend()

# Display the plot
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values for x between 0 and 2π
x = np.linspace(0, 2 * np.pi, 1000)

# Sine function without phase shift
y1 = np.sin(x)

# Sine function with phase shift of 60 degrees (π/3 radians)
y2 = np.sin(x + np.pi / 3)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label="Sine function (No phase difference)")
plt.plot(x, y2, label="Sine function (Phase difference 60°)")

# Add labels, title, and legend
# plt.title('Sine Function with and without Phase Difference')
# plt.xlabel('x (radians)')
# plt.ylabel('y')
# plt.grid(True)
# plt.legend()

# Display the plot
plt.axvline(x=1.05, color='red', linestyle='--')
plt.axhline(y=0.5, color='r', linestyle='-')
plt.show()
```
